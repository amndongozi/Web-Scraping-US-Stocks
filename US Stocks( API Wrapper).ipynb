```python
## Installing the API Wrapper
# Instead of downloading DOW data from the Yahoo Finance Website, we can install an API wrapper (yfinance) in Anaconda which is more scalable for data access
# This is done in Anaconda prompt: pip install yfinance
```

```python
import pandas as pd
import yfinance as yf
```

```python
yf.download("^DJI", multi_level_index=False) # The multi level index removes the unnecessary ^DJI duplicated row
```

```python
dowdata= yf.download("^DJI", start = '2007-01-01', end = '2020-03-31', multi_level_index=False)
dowdata
```

```python
dowdata.info()
```

```python
dowdata.to_csv("dowdata.csv")
```

```python
ticker = pd.read_csv("ticker_list")
# print(type(ticker))
ticker = ticker['Ticker'].tolist()  # Convert the ticker dataframe to a list
ticker
# print(type(ticker))
```

```python
# This comibnes the prices and the 30 ticker list of companies
prices = yf.download(ticker, start = '2007-01-01', end = '2020-03-31')
prices
```

```python
prices
```

```python
# Getting only the Closing prices for the end of the day
prices = prices.loc[:, "Close"].copy()
```

```python
prices
```

```python
prices.info()
```

```python
prices.to_csv("stock_prices.csv")
```
